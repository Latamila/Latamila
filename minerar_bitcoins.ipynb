```python
from pycoingecko import CoinGeckoAPI
```

```python
cg= CoinGeckoAPI()
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency= 'usd', days=30)
```

```python
bitcoin_data
```

```python
import pandas as pd
```

```python
data= pd.DataFrame(bitcoin_data)
```

```python
data
```

```python
bitcoin_price_data = bitcoin_data['prices']
```

```python
bitcoin_price_data
```

```python
data= pd.DataFrame(bitcoin_price_data, columns=['Tempo','Preo'])
```

```python
data
```

```python
data['Data'] = pd.to_datetime(data['Tempo'], unit='ms')
```

```python
data
```

```python
candlestick_data = data.groupby(data.Data.dt.date).agg({'Preo': [ 'min','max','first','last']})
```

```python
candlestick_data
```

```python
import plotly.graph_objects as go
```

```python
from datetime import datetime
```

```python
fig = go.Figure(data=[go.Candlestick(x=candlestick_data.index,
                open=candlestick_data['Preo']['first'],
                high=candlestick_data['Preo']['max'],
                low=candlestick_data['Preo']['min'],
                close=candlestick_data['Preo']['last'])])
```

```python
fig.show()
```
