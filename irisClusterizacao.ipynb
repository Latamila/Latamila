```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

```python
from sklearn.datasets import load_iris,load_breast_cancer,load_diabetes, load_digits,load_files
iris = load_iris()
breast = load_breast_cancer
diabetes = load_diabetes
digits = load_digits
files = load_files
```

```python
df_iris = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
sns.scatterplot(x='sepal length (cm)', y = 'petal length (cm)',
                data = df_iris, s = 60, color = 'green')
```

```python
X = df_iris[['sepal length (cm)', 'petal length (cm)']].values
```

```python
X
```

```python
#importando o metodo k-means
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters = 3, random_state = 42)
```

```python
modelo.fit(X)
```

```python
centroides = modelo.cluster_centers_
centroides
```

```python
modelo.labels_
```

```python
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=df_iris,s=60, hue=modelo.labels_, palette=['green', 'blue', 'orange'])

sns.scatterplot(x=centroides[:,0],
                y=centroides[:,1],
                marker='*',
                s=400,
                color='red')
```

```python
modelo.inertia_
```

```python
#metodo elbow
distorcoes = []

for k in range(2,11):
  KMeans_modelo = KMeans(n_clusters=k, random_state=42)
  KMeans_modelo.fit(X)
  distorcoes.append(KMeans_modelo.inertia_)

plt.plot(range(2, 11), distorcoes, marker='o')
plt.xlabel('Numero de Grupos (K)')
plt.ylabel('DistorÃ§Ã£o')
```

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise = 0.1)
df_amostra = pd.DataFrame(X,y)
df_amostra = df_amostra.rename(columns = {0: 'X1', 1: 'X2'})

df_amostra.head()
```

```python
sns.scatterplot(x='X1', y='X2',
                data = df_amostra, s=60, color='green')
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan_modelo1 = DBSCAN()
```

```python
dbscan_modelo1.fit(X)
```

```python
sns.scatterplot(x='X1', y = 'X2', data=df_amostra, c=dbscan_modelo1.labels_)
```

```python
dbscan_modelo2 = DBSCAN(eps=0.1, min_samples=8)
dbscan_modelo2.fit(X)
```

```python
sns.scatterplot(x='X1', y='X2', data=df_amostra, c= dbscan_modelo2.labels_)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nearest_neighbors = NearestNeighbors(n_neighbors=11)
neighbors = nearest_neighbors.fit(df_amostra)

distances, indices = neighbors.kneighbors(df_amostra)
distances = np.sort(distances[:,10], axis=0)

fig = plt.figure(figsize=(5,5))
plt.plot(distances)
plt.xlabel('Pontos')
plt.ylabel('Distancia')
```

```python
dbscan_modelo3 = DBSCAN(eps=0.15, min_samples=8)
dbscan_modelo3.fit(X)
```

```python
sns.scatterplot(x='X1', y='X2', data=df_amostra,c=dbscan_modelo3.labels_)
```
