Quando usamos o método pop(), estamos, literalmente, extraindo a 
coluna do dataframe. Assim, a coluna “classe”, a nossa variável alvo, será armazenada 
em uma variável chamada “classe” e será eliminada do dataframe “credito”.

```python
#extraindo as classes do dataframe

classes= np.array(credito.pop('classe'))
```

O algoritmo que utilizaremos só funciona com dados 
numéricos, e você deve ter percebido que temos muitos dados não numéricos no 
nosso conjunto. Dessa forma, não conseguiremos criar o modelo. O pacote Pandas nos oferece um método para converter dados não numéricos em dados numéricos, o método get_dummies(). O método usa uma técnica 
conhecida como one-hot encoding para criar colunas adicionais no dataframe com valores binários.


```python
credito = pd.get_dummies(credito)
```

```python
credito
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
treino,teste,treino_classes,teste_classes = train_test_split(credito, classes,
                                                             stratify= classes,
                                                             test_size=0.2,
                                                             random_state=123)
```

```python
#carregando o pacote para o modelo de arvore de decisao 
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore=DecisionTreeClassifier(random_state=123)
```

```python
arvore.fit(treino, treino_classes)
```

```python
#previsoes
previsoes_teste = arvore.predict(teste)
```

```python
#carregando os pacotes para avaliacao do modelo
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
#matriz de confusao
confusion_matrix(previsoes_teste, teste_classes)
```

A matriz de confusão (confusion_matrix) é uma das abordagens mais utilizadas 
para avaliar modelos de classificação. Na diagonal principal da matriz (103 e 
29), estão os acertos do modelo. Perceba que usamos a função confusion_matrix 
com os parâmetros: previsoes_teste e teste_classes, ou seja, as previsões feitas pelo 
modelo e as classes reais do conjunto de dados. Assim, é possível saber quando 
o modelo acertou ou errou em suas previsões. 

**Nesse caso específico, o modelo acertou 132 previsões (103 + 29)**

A diagonal 
secundária apresenta os erros do modelo: 68 (37 + 31). Podemos, assim, calcular a 
precisão do nosso modelo, simplesmente, dividindo os acertos pelo total de previsões: 132 / 200 = 0,66 (66%). A precisão total do modelo é de 66%. Se avaliarmos 
somente a precisão total, diríamos que temos um modelo promissor.

**a função 
classification_report nos fornece um relatório mais detalhado do desempenho 
do modelo**

```python
print(classification_report(teste_classes, previsoes_teste))
```

Preste atenção nos valores em destaque! A primeira linha mostra informações 
sobre a classe 0. Você deve estar lembrado(a) de que substituímos os valores “no” 
(não) para 0 e os valores “yes” (sim) para 1. 

a classe 0 representa os pareceres negativos e, nesse caso, o modelo teve 77% de precisão quando previu que 
a classe seria 0. Já para a classe 1, a precisão foi de 44%. A seta vermelha destaca 
a precisão total do modelo.

O modelo até se comporta bem quando faz previsões para a classe 0, mas 
tem um desempenho muito ruim quando a classe é 1

O modelo é muito mais preciso para a classe 
0, porque “aprendeu” mais sobre essa classe. 

O algoritmo random 
forest é uma evolução da árvore de decisão. Em linhas gerais, ele trabalha com 
várias árvores de forma simultânea, fazendo combinações diferentes de dados 
e variáveis — random forest: floresta randômica (GRUS, 201). Essa abordagem, 
em muitos casos, consegue construir modelos mais precisos. Verifiquemos! 

```python
#carregando o pacote do classificador random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier(n_estimators=100,
                           random_state=123,
                           max_features='sqrt',
                           n_jobs=1,
                           verbose=1)
```

```python
rf.fit(treino,treino_classes)
previsoes_teste_rf = rf.predict(teste)
```

Não se preocupe com a quantidade de parâmetros na criação do modelo. Veja, 
porém, que temos um parâmetro chamado n_estimators = 100. Esse parâmetro 
define, basicamente, a quantidade de árvores que o modelo manipulará. Isso é 
muito mais eficiente do que trabalhar com apenas uma única árvore. Será mesmo? Avaliaremos o resultado das previsões com o novo modelo

```python
confusion_matrix(previsoes_teste_rf, teste_classes)
```

```python
print(classification_report(teste_classes, previsoes_teste_rf))
```

**Veja que a precisão total do modelo aumentou para 76%. As precisões das classestambém melhoraram. Temos, agora, para a classe 0, uma precisão de 79% e, para a classe 1, uma precisão de 64%, ou seja, um aumento de 20 pontos percentuais 
em relação ao modelo anterior**

O que acabamos de fazer é uma estratégia comumente 
adotada, ou seja, construir modelos com abordagens diferentes usando o 
mesmo conjunto de dados. Durante a etapa de construção dos modelos, é provável que você construa várias versões, com o objetivo de melhorar os resultados.
