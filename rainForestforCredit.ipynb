Quando usamos o mÃ©todo pop(), estamos, literalmente, extraindo a 
coluna do dataframe. Assim, a coluna â€œclasseâ€, a nossa variÃ¡vel alvo, serÃ¡ armazenada 
em uma variÃ¡vel chamada â€œclasseâ€ e serÃ¡ eliminada do dataframe â€œcreditoâ€.

```python
#extraindo as classes do dataframe

classes= np.array(credito.pop('classe'))
```

O algoritmo que utilizaremos sÃ³ funciona com dados 
numÃ©ricos, e vocÃª deve ter percebido que temos muitos dados nÃ£o numÃ©ricos no 
nosso conjunto. Dessa forma, nÃ£o conseguiremos criar o modelo. O pacote Pandas nos oferece um mÃ©todo para converter dados nÃ£o numÃ©ricos em dados numÃ©ricos, o mÃ©todo get_dummies(). O mÃ©todo usa uma tÃ©cnica 
conhecida como one-hot encoding para criar colunas adicionais no dataframe com valores binÃ¡rios.


```python
credito = pd.get_dummies(credito)
```

```python
credito
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
treino,teste,treino_classes,teste_classes = train_test_split(credito, classes,
                                                             stratify= classes,
                                                             test_size=0.2,
                                                             random_state=123)
```

```python
#carregando o pacote para o modelo de arvore de decisao 
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore=DecisionTreeClassifier(random_state=123)
```

```python
arvore.fit(treino, treino_classes)
```

```python
#previsoes
previsoes_teste = arvore.predict(teste)
```

```python
#carregando os pacotes para avaliacao do modelo
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
#matriz de confusao
confusion_matrix(previsoes_teste, teste_classes)
```

A matriz de confusÃ£o (confusion_matrix) Ã© uma das abordagens mais utilizadas 
para avaliar modelos de classificaÃ§Ã£o. Na diagonal principal da matriz (103 e 
29), estÃ£o os acertos do modelo. Perceba que usamos a funÃ§Ã£o confusion_matrix 
com os parÃ¢metros: previsoes_teste e teste_classes, ou seja, as previsÃµes feitas pelo 
modelo e as classes reais do conjunto de dados. Assim, Ã© possÃ­vel saber quando 
o modelo acertou ou errou em suas previsÃµes. 

**Nesse caso especÃ­fico, o modelo acertou 132 previsÃµes (103 + 29)**

A diagonal 
secundÃ¡ria apresenta os erros do modelo: 68 (37 + 31). Podemos, assim, calcular a 
precisÃ£o do nosso modelo, simplesmente, dividindo os acertos pelo total de previsÃµes: 132 / 200 = 0,66 (66%). A precisÃ£o total do modelo Ã© de 66%. Se avaliarmos 
somente a precisÃ£o total, dirÃ­amos que temos um modelo promissor.

**a funÃ§Ã£o 
classification_report nos fornece um relatÃ³rio mais detalhado do desempenho 
do modelo**

```python
print(classification_report(teste_classes, previsoes_teste))
```

O algoritmo random 
forest Ã© uma evoluÃ§Ã£o da Ã¡rvore de decisÃ£o. Em linhas gerais, ele trabalha com 
vÃ¡rias Ã¡rvores de forma simultÃ¢nea, fazendo combinaÃ§Ãµes diferentes de dados 
e variÃ¡veis â€” random forest: floresta randÃ´mica (GRUS, 201). Essa abordagem, 
em muitos casos, consegue construir modelos mais precisos. Verifiquemos! 

```python
#carregando o pacote do classificador random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rf= RandomForestClassifier(n_estimators=100,
                           random_state=123,
                           max_features='sqrt',
                           n_jobs=1,
                           verbose=1)
```

```python
rf.fit(treino,treino_classes)
previsoes_teste_rf = rf.predict(teste)
```

NÃ£o se preocupe com a quantidade de parÃ¢metros na criaÃ§Ã£o do modelo. Veja, 
porÃ©m, que temos um parÃ¢metro chamado n_estimators = 100. Esse parÃ¢metro 
define, basicamente, a quantidade de Ã¡rvores que o modelo manipularÃ¡. Isso Ã© 
muito mais eficiente do que trabalhar com apenas uma Ãºnica Ã¡rvore. SerÃ¡ mesmo? Avaliaremos o resultado das previsÃµes com o novo modelo

```python
confusion_matrix(previsoes_teste_rf, teste_classes)
```

```python
print(classification_report(teste_classes, previsoes_teste_rf))
```

**Veja que a precisÃ£o total do modelo aumentou para 76%. As precisÃµes das classestambÃ©m melhoraram. Temos, agora, para a classe 0, uma precisÃ£o de 79% e, para a classe 1, uma precisÃ£o de 64%, ou seja, um aumento de 20 pontos percentuais 
em relaÃ§Ã£o ao modelo anterior**

O que acabamos de fazer Ã© uma estratÃ©gia comumente 
adotada, ou seja, construir modelos com abordagens diferentes usando o 
mesmo conjunto de dados. Durante a etapa de construÃ§Ã£o dos modelos, Ã© provÃ¡vel que vocÃª construa vÃ¡rias versÃµes, com o objetivo de melhorar os resultados.
