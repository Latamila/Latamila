# TENDÊNCIA

Na análise de série temporal, uma tendência refere-se à direção geral dos dados ao longo do tempo, como uma tendência ascendente ou uma tendência descendente. Uma tendência pode ser de longo prazo ou de curto prazo.As  tendências  podem  ser  observadas  em  dados  de  séries  temporais  observando  um gráfico  dos  dados  ao  longo  do  tempo.  Por  exemplo,  se  os  dados  aumentam  ou  diminuem consistentemente ao longo do tempo, diz-se que eles têm uma tendência positiva ou negativa, respectivamente. Se os dados são relativamente estáveis e exibem pouca mudança ao longo do tempo, diz-se que tem uma tendência plana.As tendências podem ser causadas por uma variedade de fatores, como mudanças no sistema ou processo que está sendo medido, mudanças em fatores externos, como condições econômicas  ou  inovações  tecnológicas,  ou  mudanças  no  comportamento  dos  indivíduos  ou organizações que estão sendo estudadas.A análise de tendências é uma parte importante da análise de séries temporais, pois permite que os analistas entendam os padrões e tendências nos dados e façam previsões sobre valores futuros. Existem várias técnicas estatísticas que podem ser usadas para identificar e quantificar  tendências  em  dados  de  séries  temporais,  como  médias  móveis,  suavização exponencial e regressão linear.

# SAZONALIDADE

Na análise de séries temporais, a sazonalidade refere-se a padrões regulares e repetidos nos dados que ocorrem em um período de tempo específico, como diariamente, semanalmente, mensalmente ou anualmente. A sazonalidade pode ser causada por uma variedade de fatores, como mudanças no clima, comportamento do consumidor ou condições econômicas.Por exemplo, as vendas de roupas de inverno podem ser maiores nos meses de inverno devido ao clima mais frio, enquanto as vendas de sorvete podem ser maiores no verão devido ao clima mais quente. Da mesma forma, as vendas no varejo podem ser maiores na temporada de festas devido ao aumento dos gastos do consumidor.A sazonalidade pode ser observada em dados de séries temporais observando um gráfico dos dados ao longo do tempo. Se os dados exibem padrões regulares e repetidos durante um período de tempo específico, dizemos que são sazonais. A sazonalidade pode ser quantificada usando técnicas estatísticas como decomposição ou análise espectral.Compreender a sazonalidade é importante na análise de séries temporais porque pode afetar a precisão das previsões e a interpretação das tendências nos dados. Por exemplo, se uma série  temporal  apresentar  forte  sazonalidade,  pode  ser  necessário  remover  o  componente sazonal antes de analisar a tendência ou fazer previsões. Isso pode ser feito usando técnicas como decomposição sazonal ou ajuste sazonal.

# ESTACIONARIEDADE

A estacionariedade é um importante conceito na modelagem de séries temporais e é caracterizada por uma variável que se comporta de forma aleatória ao longo do tempo ao redor de uma média constante.A estacionariedade  se  refere  à  propriedade  de  uma  série  temporal  na  qual  as propriedades estatísticas (como média, variância e autocorrelação) são constantes ao longo do tempo.  Uma  série  temporal  estacionária  é  aquela  que  exibe  pouca  ou  nenhuma  tendência, sazonalidade ou mudanças nas propriedades estatísticas ao longo do tempo.Existem vários tipos de estacionariedade na análise de séries temporais:Estacionária estrita: Uma série temporal é estritamente estacionária se a distribuição de probabilidade dos valores permanecer a mesma ao longo do tempo. Isso significa que a média, a variância e a autocorrelação da série não mudam com o tempo.Estacionária fraca: Uma série temporal é fracamente estacionária se a média e a variância da série não mudam com o tempo, mas a autocorrelação pode variar.Estacionáriadetendência: Ocorre  quando  uma  série  temporal exibeuma  tendência constante ao longo do tempo, mas a média e a variância podem mudar.Estacionária sazonal: Uma série temporal é estacionária sazonal se exibir um padrão de sazonalidade constante ao longo do tempo, mas a média e a variância podem mudar.A  estacionariedade  é  um  conceito  importante  na  análise  de  séries  temporais  porque muitas técnicas estatísticas, como previsão e teste de hipóteses, são baseadas na suposição de que  os  dados  são  estacionários.  Se  uma  série  temporal  for  não  estacionária  (ou  seja,  exibir tendências, sazonalidade ou mudanças nas propriedades estatísticas ao longo do tempo), pode ser necessário transformar os dados ou remover os componentes não estacionários antes de aplicar essas técnicas.Data Science Academy camilaaeromoca@gmail.com 61232b50e32fc3d16d5bf43b

# METODOS DE CLÁSSICOS

Existem  vários métodos  clássicos  que  são  comumente  usados  na  análise  de  séries temporais. Esses métodos não são classificados como Inteligência Artificial (IA), mas podem ser muito úteis em situações em que a explicabilidade do processo de análise ou previsão é tão importante  quanto  o  resultado  da  análise  ou  da  previsão.  Técnicas  baseadas  em  IA  podem oferecer ótimos resultados, mas a explicabilidade pode ser limitada.Este curso tem como objetivo principal trazer para você técnicas modernas e avançadas de  Inteligência  Artificial  para  trabalhar  com  séries  temporais,  mas  uma  compreensão  dos métodos  clássicos  vai  dar  a  você  os  fundamentos  necessários  para  compreender  como  as técnicas de IA funcionam. Aqui estão osprincipaismétodos clássicos de análise de séries temporais:

#Modelo  Autorregressivo  (AR-AutoRegressive)

Um  modelo  autorregressivo  é  um modelo estatístico que explica uma série temporal em termos de seus valores passados. Um modelo AR assume que o valor atual da série temporal é uma função linear de seus valores passados, com um conjunto de parâmetros que determinam a força do relacionamento.

#Modelo  de  Média  Móvel  (MA-Moving  Average)

Um modelo de média móvel é um modelo estatístico que explica uma série temporal em termos de erros ou resíduos a partirdamédia móvel. Um modelo MA assume que o valor atual da série temporal é uma função linear dos erros ou resíduos damédia móvel, com um conjunto de parâmetros que determinam a força da relação.

#Modelo   de Média Móvel Integrada Autorregressiva   (ARIMA-AutoRegressive Integrated  Moving  Average)

Um  modelo  de  média  móvel  integrada  autorregressiva  é  um modelo estatístico que combina os modelos autorregressivo (AR) e de média móvel (MA). Um modelo ARIMA pode ser usado para analisar e prever dados de séries temporais que exibem tendências  e  sazonalidade.O  modelo  ARIMA  tem  duas importantes variações:  SARIMA  e ARIMAX.

#Suavização Exponencial(Exponential  Smoothing)

A  suavização  exponencial  é  uma técnica  estatística  que  envolve  ponderar  os  dados  em  uma  série  temporal  para  que  as observações mais recentes tenham maior influência no valor médio. A suavização exponencial é frequentemente usada para prever valores futuros de uma série temporal com base em seus valores passados.

Já os segundos usam uma combinação linear de valores passados da própria variável.Logo, embora até possa ser usada, a Regressão Linear não é o método ideal para analisar e prever séries temporais.Vejamos abaixo alguns casos especiais do modelo autorregressivo(AR):

# AR(0): Ruído Branco

Se definirmos o parâmetro p como zero (AR(0)), sem termos autorregressivos, esta série temporal é apenas ruído branco. Cada ponto de dados é amostrado de uma distribuição com uma média de 0 e umavariância de sigma ao quadrado. Isso resulta em uma sequência de números aleatórios que não podem ser previstos. Isso é realmente útil, pois pode servir como uma hipótese nula e proteger nossas análises de aceitar padrões falsos positivos.

#AR(1): Passeios aleatórios e oscilações

Com o parâmetro p definido como 1, estamos levando em consideração o registro de data e hora anterior ajustado por um multiplicador e, em seguida, adicionando ruído branco. Se o multiplicador  for  0,  obtemos  ruído  branco  e,  se  o  multiplicador  for  1,  obtemos  um  passeio aleatório. Se o multiplicador estiver entre 0 < α₁ < 1, então a série temporal apresentará reversão à média. Isso significa que os valores tendem a girar em torno de 0 e reverter para a média após regredir a partir dela.

#AR(p): Termos de ordem superior

Aumentar ainda mais o parâmetro p significa apenas voltar mais longe e adicionar mais timestampsajustados  por  seus  próprios  multiplicadores.  Podemos  retroceder  o  quanto quisermos,  mas  à  medida  que  recuamos,  é  mais  provável  que  devamos  usar  parâmetros adicionais, como a média móvel (MA(q)).

# MODELOS DE MÉDIA MÓVEL

MA(q)é  o  modelo  de  média  móvel  e  q  é  o  número  de  termos  de  erro  de  previsão defasados na previsão. Em um modelo MA(1), nossa previsão é um termo constante mais o termo de ruído branco anterior vezes um multiplicador, adicionado ao termo de ruído branco atual. Isso é apenas probabilidade + estatística simples, pois estamos ajustando nossa previsão com base nos termos de ruído branco anteriores.


# Modelos Autorregressivos Integrados com Média Móvel (ARIMA)

Os  modelos  autorregressivos  integrados  com  média  móvel  (ARIMA)  são  modelos estatísticos que combinam os modelos autorregressivos (AR) e média móvel (MA) para explicar e prever uma série temporal. O modelo ARIMA é um modelo linear que pode ser usado paraanalisar e prever séries temporais que apresentam tendências e sazonalidade. A forma geral de um modelo ARIMA é dada pela seguinte equação:
# y(t) = c + Σφ(i) * y(t-i) + Σθ(i) * ε(t-i) + ε(t)

onde y(t) é o valor atual da série temporal no tempo t, c é um termo constante, φ(i) é o coeficiente autorregressivo para o i-ésimo atraso (lag), θ(i) é o coeficiente de média móvel para o i-ésimo atraso, y(t-i) é o valor da série temporal no i-ésimo atraso, ε(t-i) é o erro de média móvel no i-ésimo atraso e ε(t) é umtermo de erro de ruído branco.Um modelo ARIMA não sazonal é representado pela expressão:ARIMA(p,d,q), onde:•p é o número de termos autorregressivos.•d é o número de diferenças não sazonais necessárias para estacionariedade.•q é o número de erros de previsão defasados na equação de previsão.ARIMA  é  um  processo  univariado.  Os  valores  atuais  de  uma  série  de  dados  são correlacionados aos valores passados na mesma série para produzir o componente AR, também conhecido como p. Os valores atuais de uma condição de erro aleatória são correlacionados aos valores passados para produzir o componente MA, q. Os valores de média e variância dos dados passados e atuais são considerados como estacionários e inalterados ao longo do tempo. Se necessário,  um  componente  I  (simbolizado  pelo  d)  será  adicionado  para  corrigir  a  falta  de estacionariedade por meio de diferenciação.

# Funcionamento de Modelos Autorregressivos Integrados com Média Móvel Sazonal (SARIMA)

Um modelo SARIMA (Seasonal Autoregressive Integrated Moving Average) é um modelo estatístico que combina os modelos autorregressivos integrados com média móvel (ARIMA) com um  componente  sazonal  para  explicar  e  prever  uma  série temporal.  O  componente  sazonal reflete a influência de padrões sazonais repetitivos, como a estação do ano, nas observações da série temporal.Sazonalidade em séries temporais são padrões regulares que se repetem ao longo do tempo, onde S define o número de períodos até que o padrão volte a se repetir. A presença de sazonalidade pode fazer com que a série não seja estacionária, sendo necessário modelar esse componente para que sejamos capazes de trabalhar com a série.Cada negócio tem sua própria forma desazonalidade. Ela pode ser diária, pode ser anual (imagine um negócio que tem picos de vendas durante eleições ouCopa do Mundo), ou pode ser até por hora. Quando falamos em previsão de séries temporais, há uma forma de lidarmos com isso: usando o modelo SARIMA.A  sigla  SARIMA  tem  o  mesmo  significado  que  ARIMA,  onde  o  “S”  representa  a sazonalidade. Seu processo de construção é exatamente o mesmo para um modelo ARIMA, onde você  terá  que  verificar  se  sua  série  é  estacionária,  usar  as  funções  ACF  e  PACF  para ver  a quantidade de termos, etc... (veremos isso no Estudo de Caso). Sua principal diferença está no momento  de  treinar  o  modelo,  onde  você  usará  uma  nova  função  e  irá  adicionar  mais  um argumento.O modelo Seasonal ARIMA (SARIMA) permite tal abordagem, pois incorpora tanto os fatores não-sazonais, quanto os fatores sazonais em um modelo. Esta é a representação de um modelo SARIMA:SARIMA(p,d,q)x(P,D,Q,s)onde p, d, q são as ordens e diferença do modelo não sazonal e P, D, Q seu pares para o modelo sazonal, destaca-se que S é o intervalo de tempo que o padrão sazonal se repete.

## Estacionariedade

Vamos criar uma função para visualizar a estacionariedade.

```python
# Função para visualizar estacionaridade
def visualiza_estacionaridade(dados_serie):

    # Calcula média e desvio padrão móveis
    rolling_mean = dados_serie.rolling(12).mean()
    rolling_std = dados_serie.rolling(12).std()

    # Plot
    fig = plt.figure(figsize = (12,6))
    time_series = plt.plot(dados_serie, label = 'Itens Vendidos')
    mean = plt.plot(rolling_mean, color = 'yellow', label = "Média Móvel")
    std = plt.plot(rolling_std, color = 'green', label = "Desvio Padrão Móvel")

    plt.legend(loc = 'best')
    plt.title("Visualizando a Estacionaridade")
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
# Aplica a função
visualiza_estacionaridade(dados_ts)
```

## Modelo SARIMAX

O modelo usado para previsão depende dos dados da série temporal. A série temporal precisa ser dividida para ver se contém ou não partes autorregressivas (AR) e partes de média móvel (MA), integração e sazonalidade.

Os gráficos de função de autocorrelação (ACF) e os gráficos de função de autocorrelação parcial (PACF) podem ser usados para determinar se a série temporal contém partes AR e MA.

Parâmetros que devem ser deifnidos para o modelo SARIMAX:

* p - número de atrasos ou lags (AR)
* d - número de diferenciações necessárias
* q - número de atrasos de erro (MA)
* P - número de atrasos sazonais (AR)
* D - número de diferenças sazonais necessárias
* Q - número de defasagens de erro sazonal (MA)
* S - sazonalidade

### Plots de Autocorrelação (ACF & PACF)

* Há forte correlação em lag = 1, 7, 14, 21, 28, 35 para o gráfico ACF. Ocorre a cada 7 lags.


* Há forte correlação em lag = 1, 2, 3, 4, 5, 6, 13, 20, 27 para gráfico PACF. Ocorre em 6, seguido por atrasos a cada 7, mas diminui.

Previsão $y(t)$

## ATRIBUTOS DATE TIME

Podemos usar o registro de data e hora e gerar um recurso chamado *is_holiday*, que informa se um dia é feriado ou não. Além disso podemos transformar a data e hora em diversas outras variáveis de data.

```python

# Função para listar feriados
def lista_holidays(df):

    df['data'] = pd.to_datetime(df.data)

    min_year = df.data.min().year
    max_year = df.data.max().year

    years_list = pd.period_range(min_year, max_year, freq = 'Y')

    list_of_holidays = []

    for year in years_list:
        list_of_holidays.append(holidays.US(years = int(str(year))).keys())

    holiday_list = [item for sublist in list_of_holidays for item in sublist]

    return holiday_list
```

```python
# Função para criar os novos atributos date e time
def cria_atributos_date_time(df):

    df['data'] = pd.to_datetime(df.data)
    df['month'] = df.data.dt.month
    df['day_of_month'] = df.data.dt.day
    df['day_of_year'] = df.data.dt.dayofyear
    df['week_of_year'] = df.data.dt.weekofyear
    df['day_of_week'] = df.data.dt.weekday + 1
    df['year'] = df.data.dt.year
    df['is_weekend'] = df.data.dt.weekday // 5
    df['start_of_month'] = df.data.dt.is_month_start.astype(int)
    df['end_of_month'] = df.data.dt.is_month_end.astype(int)
    df['is_holiday'] = np.where(df.data.isin(lista_holidays(df)), 1, 0)

    return df
```

##  ATRIBUTOS ROLLING WINDOW

Com a média móvel inserimos uma janela de tempo e calculamos a média ou a demanda média desse período de tempo.

### Atributos Exponentially Weight Mean (EWM)

Esse recurso aplica peso aos valores da série temporal. Valores mais recentes terão um peso maior aplicado a eles. Isso porque os pontos mais recentes serão mais relevantes para as previsões futuras.

```python
# Lags
lags = [91, 98, 105, 112, 180, 270, 365, 546, 728]
```

```python
# Lista de alfas
alpha_list = [0.95, 0.9, 0.8, 0.7, 0.5]
```

```python
# Função
def cria_atributos_ewm(df, alphas, lags):
    for alpha in alphas:
        for lag in lags:
            df['sales_ewm_alpha_' + str(alpha).replace(".","") +
              "_lag_" + str(lag)] = df.groupby(["loja", "produto"]).itens_vendidos.transform(
            lambda x: x.shift(lag).ewm(alpha = alpha).mean())
    return df
```

```python
df = cria_atributos_ewm(df, alpha_list, lags)
```
