```python
states = []
for cidades in cidades_minas:
  cidade = str(cidades).lower()
  cidade = cidade.replace(' ','-')
  cidade = unidecode.unidecode(str(cidade))
  print(unidecode.unidecode(str(cidade)))
  states.append(cidade)
```

```python
states
```

```python
def scrap_state_info(state: str) -> dict:

  print(f' Picking {state} info....')
  state_url= f'https://www.ibge.gov.br/cidades-e-estados/mg/{state}.html'
  page = requests.get(state_url)

  soup = BeautifulSoup(page.content, 'html.parser')
  indicadors = soup.select('.indicador')

  state_dict = {
      ind.select('.ind-label')[0].text: ind.select('.ind-value')[0].text
      for ind in indicadors
  }

  state_dict['Cidade'] = state
  
  return state_dict
```

```python
for state in states:
  print(state)
```

```python
states_data = [scrap_state_info(state) for state in states]
```

```python
cidades_tratado = cidades_df.replace(
    {'\.':'',
     ',': '.',
     '\[\d+\]':'',
     ' hab/km²': '',
     ' km²': '',
     ' pessoas': '',
     ' matrículas': '',
     ' óbitos por mil nascidos vivos': '',
     'R\$.*': '',
     '-':' ',
     ' veículos': '',
     '%': '',}, regex=True)
```
