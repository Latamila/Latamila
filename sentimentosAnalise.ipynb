```python
lista_stop = nltk.corpus.stopwords.words('portuguese')
```

```python
def aplica_stemming_stopwords(dados):
  stemmer = nltk.stem.RSLPStemmer()
  frases_processadas=[]
  for (frases,sentimento) in dados:
    com_Stemming = [str(stemmer.stem(p)) for p in frases.split() if p not in lista_stop]
    frases_processadas.append((com_Stemming, sentimento))
  return frases_processadas
```

```python
base_treinamento = sent_trein.to_records(index = False)
```

```python
import nltk
```

```python
base_treinamento
```

```python
frases_processadas_treinamento = aplica_stemming_stopwords(base_treinamento)
```

```python
pd.DataFrame(frases_processadas_treinamento, columns=['Frase','Sentimento']).sample(10)
```

```python
base_teste = sent_teste.to_records(index = False)
```

```python
frases_processadas_teste = aplica_stemming_stopwords(base_teste)
```

```python
pd.DataFrame(frases_processadas_teste)
```

```python
def busca_palavras(frases):
  todas_palavras=[]
  for (palavras, sentimento) in frases:
    todas_palavras.extend(palavras)
  return todas_palavras
```

```python
palavras_treinamento = busca_palavras(frases_processadas_treinamento)
```

```python
palavras_teste = busca_palavras(frases_processadas_teste)
```

```python
def busca_frequencia(palavras):
  palavras = nltk.FreqDist(palavras)
  return palavras
```

```python
frequencia_treinamento = busca_frequencia(palavras_treinamento)
frequencia_treinamento.most_common(10)
```

```python
frequencia_teste = busca_frequencia(palavras_teste)
frequencia_teste.most_common(10)
```

```python
def busca_palvras_unicas(frequencia):
  freq = frequencia.keys()
  return freq

palavras_unicas_treinamento = busca_palvras_unicas(frequencia_treinamento)
palavras_unicas_teste = busca_palvras_unicas(frequencia_teste)
```

```python
def extrator_palavras_treinamento(documento):
  doc = set(documento)
  caracteristicas = {}
  for palavras in palavras_unicas_treinamento:
    caracteristicas['%s' % palavras] = (palavras in doc)
  return caracteristicas
```

```python
def extrator_palavras_teste(documento):
  doc = set(documento)
  caracteristicas = {}
  for palavras in palavras_unicas_teste:
    caracteristicas['%s' % palavras] = (palavras in doc)
  return caracteristicas
```

```python
base_completa_treinamento = nltk.classify.apply_features(extrator_palavras_treinamento, frases_processadas_treinamento)
```

```python
base_completa_teste = nltk.classify.apply_features(extrator_palavras_teste, frases_processadas_teste)
```

```python
classificador = nltk.NaiveBayesClassifier.train(base_completa_treinamento)
```

```python
print(classificador.labels())
```

```python
print(classificador.show_most_informative_features(10))
```

```python
print(nltk.classify.accuracy(classificador, base_completa_teste))
```

```python
from nltk.metrics import ConfusionMatrix
```

```python
esperado=[]
previsto=[]

```

```python
for (frase,classe) in base_completa_teste:
  resultado = classificador.classify(frase)
  previsto.append(resultado)
  esperado.append(classe)

matriz = ConfusionMatrix(esperado, previsto)
print(matriz)
```

```python
teste = 'Nossa, que voo atrasado. Perdi minha conexão!'
testeStemming = []
stemmer = nltk.stem.RSLPStemmer()
for (palavras_treinamento) in teste.split():
  comStem = [p for p in palavras_treinamento.split()]
  testeStemming.append(str(stemmer.stem(comStem[0])))

novo = extrator_palavras_treinamento(testeStemming)

distribuicao = classificador.prob_classify(novo)
for classe in distribuicao.samples():
  print('%s: %f' % (classe, distribuicao.prob(classe)))
```
